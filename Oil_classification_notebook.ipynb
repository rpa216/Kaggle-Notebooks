```python
%matplotlib inline
import pandas as pd
import numpy as np
import re
```

```python
# The dataset
df=pd.read_csv("olive.csv")
df.rename(columns={df.columns[0]:'areaname'}, inplace=True)
df.areaname=df.areaname.map(lambda x: x.split('.')[-1])
df.sample(50).head(10)
```

## 1. LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import model_selection

df_lda = df.copy(deep=True)

X = np.array(df_lda.drop(['areaname','region', 'area'],1))
y = np.array(df_lda['areaname'])
z = np.array(df_lda['region'])

X_train1, X_test1, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.33)
X_train2, X_test2, z_train, z_test = model_selection.train_test_split(X,z,test_size=0.33)

ldaclf1 = LinearDiscriminantAnalysis()
ldaclf2 = LinearDiscriminantAnalysis()

ldaclf1.fit(X_train1, y_train)
ldaclf2.fit(X_train2, z_train)

areaname_accuracy = ldaclf1.score(X_test1, y_test)
print("Accuracy when predicting Areaname =", areaname_accuracy*100,"%")

region_accuracy = ldaclf2.score(X_test2, z_test)
print("Accuracy when predicting Region =", region_accuracy*100,"%")

data_for_prediction = np.array([1261,121,312,7238,877,47,65,25])
data_for_prediction = data_for_prediction.reshape(1,-1)

prediction_of_areaname = ldaclf1.predict(data_for_prediction)
print("Prediction of Areaname for the given input of fatty acid contents =" ,prediction_of_areaname)

prediction_of_region = ldaclf2.predict(data_for_prediction)
print("Prediction of Region for the given input of fatty acid contents =" ,prediction_of_region)
```

## 2. Logistic Regression

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn import model_selection

df_logReg = df.copy(deep=True)

X = np.array(df_logReg.drop(['areaname','region', 'area'],1))
y = np.array(df_logReg['areaname'])
z = np.array(df_logReg['region'])

X_train1, X_test1, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.33)
X_train2, X_test2, z_train, z_test = model_selection.train_test_split(X,z,test_size=0.33)

logregclf1 = LogisticRegressionCV()
logregclf2 = LogisticRegressionCV()

logregclf1.fit(X_train1, y_train)
logregclf2.fit(X_train2, z_train)

areaname_accuracy = logregclf1.score(X_test1, y_test)
print("Accuracy when predicting Areaname =", areaname_accuracy*100,"%")

region_accuracy = logregclf2.score(X_test2, z_test)
print("Accuracy when predicting Region =", region_accuracy*100,"%")

data_for_prediction = np.array([1261,121,312,7238,877,47,65,25])
data_for_prediction = data_for_prediction.reshape(1,-1)

prediction_of_areaname = logregclf1.predict(data_for_prediction)
print("Prediction of Areaname for the given input of fatty acid contents =" ,prediction_of_areaname)

prediction_of_region = logregclf2.predict(data_for_prediction)
print("Prediction of Region for the given input of fatty acid contents =" ,prediction_of_region)
```

## 3. SVM

```python
from sklearn import preprocessing, model_selection, svm

df_svm = df.copy(deep=True)

X = np.array(df_svm.drop(['areaname','region', 'area'],1))
y = np.array(df_svm['areaname'])
z = np.array(df_svm['region'])

#Splitting the dataset into training and testing data.
#Training dataset = 77% of the original dataset
#Testing dataset = 33% of the original datatset

#Spliting dataframe into testing and training dataset for prediction of areaname.
X_train1, X_test1, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.33)
X_train2, X_test2, z_train, z_test = model_selection.train_test_split(X,z,test_size=0.33)

svmclf1 = svm.SVC(kernel='linear')
svmclf2 = svm.SVC(kernel='linear')

#Training the model on the training dataset
svmclf1.fit(X_train1, y_train)
svmclf2.fit(X_train2, z_train)

#Testing the model on the testing dataset and calculating the accuracy.
areaname_accuracy = svmclf1.score(X_test1, y_test)
print("Accuracy when predicting Areaname =", areaname_accuracy*100,"%")

region_accuracy = svmclf2.score(X_test2, z_test)
print("Accuracy when predicting Region =", region_accuracy*100,"%")

data_for_prediction = np.array([1261,121,312,7238,877,47,65,25])
data_for_prediction = data_for_prediction.reshape(1,-1)

prediction_of_areaname = svmclf1.predict(data_for_prediction)
print("Prediction of Areaname for the given input of fatty acid contents =" ,prediction_of_areaname)

prediction_of_region = svmclf2.predict(data_for_prediction)
print("Prediction of Region for the given input of fatty acid contents =" ,prediction_of_region)
```

## 4. k-NN

```python
from sklearn import preprocessing, model_selection, neighbors
df_knn = df.copy(deep=True)

X = np.array(df_knn.drop(['areaname','region', 'area'],1))
y = np.array(df_knn['areaname'])
z = np.array(df_knn['region'])

#Splitting the dataset into training and testing data.
#Training dataset = 77% of the original dataset
#Testing dataset = 33% of the original datatset

#Spliting dataframe into testing and training dataset for prediction of areaname.
X_train1, X_test1, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.33)
X_train2, X_test2, z_train, z_test = model_selection.train_test_split(X,z,test_size=0.33)

knnclf1 = neighbors.KNeighborsClassifier(n_neighbors=5, weights='distance', algorithm='brute')
knnclf2 = neighbors.KNeighborsClassifier(n_neighbors=5, weights='uniform', algorithm='kd_tree')

#Training the model on the training dataset
knnclf1.fit(X_train1, y_train)
knnclf2.fit(X_train2, z_train)

#Testing the model on the testing dataset and calculating the accuracy.
areaname_accuracy = knnclf1.score(X_test1, y_test)
print("Accuracy when predicting Areaname =", areaname_accuracy*100,"%")

region_accuracy = knnclf2.score(X_test2, z_test)
print("Accuracy when predicting Region =", region_accuracy*100,"%")

#data_for_prediction = np.array([827, 20, 100, 5656, 900, 21, 90, 15])
data_for_prediction = np.array([1261,121,312,7238,877,47,65,25])
data_for_prediction = data_for_prediction.reshape(1,-1)

prediction_of_areaname = knnclf1.predict(data_for_prediction)
print("Prediction of Areaname for the given input of fatty acid contents =" ,prediction_of_areaname)

prediction_of_region = knnclf2.predict(data_for_prediction)
print("Prediction of Region for the given input of fatty acid contents =" ,prediction_of_region)
```

## 5. Random Forest

```python
from sklearn import preprocessing, model_selection
from sklearn.ensemble import RandomForestClassifier

df_randFor = df.copy(deep=True)

X = np.array(df_randFor.drop(df_randFor[['areaname','region','area']],1))
y = np.array(df_randFor['areaname'])
z = np.array(df_randFor['region'])

#Splitting the dataset into training and testing data.
#Training dataset = 77% of the original dataset
#Testing dataset = 33% of the original datatset

#Spliting dataframe into testing and training dataset for prediction of areaname.
X_train1, X_test1, y_train, y_test = model_selection.train_test_split(X,y,test_size=0.33,random_state=4)

#Spliting dataframe into testing and training dataset for prediction of region.
X_train2, X_test2, z_train, z_test = model_selection.train_test_split(X,z,test_size=0.33,random_state=4)

rf1 = RandomForestClassifier(n_estimators=500)
rf2 = RandomForestClassifier(n_estimators=500)

#Training the model on the training dataset
rf1.fit(X_train1, y_train)
rf2.fit(X_train2, z_train)

#Testing the model on the testing dataset and calculating the accuracy.
areaname_accuracy = rf1.score(X_test1, y_test)
print("Accuracy of predicting Areaname =", areaname_accuracy*100,"%")

region_accuracy = rf2.score(X_test2, z_test)
print("Accuracy of predicting Region =", region_accuracy*100,"%")

data_for_prediction = np.array([1271,101,512,9238,77,487,45,105])
data_for_prediction = data_for_prediction.reshape(1,-1)
prediction1 = rf1.predict(data_for_prediction)
print("Areaname for the given sample of fatty acid contents =" ,prediction1)

prediction = rf2.predict(data_for_prediction)
print("Region for the given sample of fatty acid contents =" ,prediction)
```
