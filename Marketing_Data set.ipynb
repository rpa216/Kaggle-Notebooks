```python
def prediction_data(A,B):
    a = len(A.columns)
#     if a == 21:
#         #print("test file is ok.. we can go ahead with prediction")
#     else:
#         print("The test_file needs the required data")
    #converting the nominal attribute to categorical attributes and then label encoding it
    A['education_encoded'] = A['education'].astype('category').cat.codes
    A['job_encoded'] = A['job'].astype('category').cat.codes
    A['marital_encoded'] = A['marital'].astype('category').cat.codes
    A['housing_encoded']=A['housing'].astype('category').cat.codes
    A['loan_encoded']=A['loan'].astype('category').cat.codes
    A['contact_encoded']=A['contact'].astype('category').cat.codes
    A['poutcome_encoded']=A['poutcome'].astype('category').cat.codes
    A['day_of_week_encoded'] = A['day_of_week'].astype('category').cat.codes
    A['month_encoded']=A['month'].astype('category').cat.codes
    A['y_encoded'] = A['y'].astype('category').cat.codes
    a1 = len(A.columns)
    X = np.array(A.drop(['job','education', 'marital','default', 'housing', 'contact','loan', 'poutcome','day_of_week', 'month','y','y_encoded'],1))
    y = np.array(A.y_encoded.tolist())
    X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.9,random_state=12)
    B['education_encoded'] = B['education'].astype('category').cat.codes
    B['job_encoded'] = B['job'].astype('category').cat.codes
    B['marital_encoded'] =B['marital'].astype('category').cat.codes
    B['housing_encoded']=B['housing'].astype('category').cat.codes
    B['loan_encoded']=B['loan'].astype('category').cat.codes
    B['contact_encoded']=B['contact'].astype('category').cat.codes
    B['poutcome_encoded']=B['poutcome'].astype('category').cat.codes
    B['day_of_week_encoded'] = B['day_of_week'].astype('category').cat.codes
    B['month_encoded']=B['month'].astype('category').cat.codes
    X_1 = np.array(B.drop(['job','education', 'marital','default', 'housing', 'contact','loan', 'poutcome','day_of_week', 'month'],1))
    clf = SVC(kernel='linear').fit(X_train, y_train)
    accuracy = clf.score(X_test, y_test)
    #print("measured_accuracy:", accuracy*100)
    scores = cross_val_score(clf, X_train, y_train, cv=10)
    #print(scores)
    example_measure = np.array(X_1)
    x=len(example_measure)
    example_measure = example_measure.reshape(x,-1)
    prediction = clf.predict(example_measure)
    return(prediction)
```
